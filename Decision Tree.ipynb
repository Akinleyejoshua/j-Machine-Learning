## Introduction to the datasets
the heart diesase datasets was gotten from a github ropository https://github.com/Akinleyejoshua/j-ml-projects/tree/main/Datasets, https://github.com/Akinleyejoshua/j-ml-projects/blob/main/Datasets/heart-disease.csv

from the datasets data below we can see that the feature sets is `13` and target set is `1`

```python
import pandas as pd
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease.shape
```

```python
# now we have imported the data let us discibe the data 
heart_disease.describe()
# from the data we can see the count, mean, minumum value, max value, standard deviation, 25 percentile ...
```

```python
# select the independent feature columns into X variable and y as the dependent variable (target)
X = heart_disease.drop("target", axis=1)
y = heart_disease["target"]
```

```python
# Now we have to split our data into train and test set
# the train set is used to fit the data into the model while the test set used to test the accuracy of the data
# the test size would take about 20% or 0.2 while train 80%
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# we can instantiate the model from the decision tree
# we can import the model from sklearn
# Since this is a classification project because we are classifying if a patient has heart disease or not
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
# fit in the train set
model.fit(X_train, y_train)
# we score the model accuracy from the test set
model.score(X_test, y_test)
```

```python
# wow we have an accuracy of abour 74%, which means our model performed well on the datasets
# now let start predicting new data from the model from the test set
# then we compare the predicted results with the real result to check if it predicted it accurately
# we can see that True means predicted correct while False goes the other way round
y_pred = model.predict(X_test)
y_pred == y_test
```

### Analysis
now we can see from our result above that the model had an accuracy of about 74% which means the model learned the patterns well from the data given to it
we then made predictions from the `test` set and compared the predicted value to the actual value
